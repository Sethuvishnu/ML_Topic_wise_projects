```python
data1=data[data.Cholesterol<=(data.Cholesterol.mean()+3*data.Cholesterol.std())]
```

```python
data.shape ,data1.shape
```

```python
data2=data1[data1.Oldpeak<=(data1.Oldpeak.mean()+3*data1.Oldpeak.std())]
data2
```

```python
data3 = data2[data2.RestingBP<=(data2.RestingBP.mean()+3*data2.RestingBP.std())]
data4=data3.copy()
```

```python
data4.ExerciseAngina.replace({'N':0,'Y':1},inplace=True)
```

```python
data4.ST_Slope.replace(
    {
        'Down': 1,
        'Flat': 2,
        'Up': 3
    },
    inplace=True
)
```

```python
data4.RestingECG.replace(
    {
        'Normal': 1,
        'ST': 2,
        'LVH': 3
    },
    inplace=True)
```

```python
data5=pd.get_dummies(data4,drop_first=True)
```

```python
data5
```

```python
x=data5.drop('HeartDisease',axis='columns')
y=data5['HeartDisease']
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_=scaler.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(x_, y, test_size=0.2, random_state=30)
```

```python
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier()
model.fit(X_train,y_train)
model.score(X_test,y_test)

```

```python
from sklearn.decomposition import PCA
pca=PCA(0.95)
x_pca=pca.fit_transform(x)
```

```python
X_train_pca, X_test_pca, y_train, y_test = train_test_split(x_pca, y, test_size=0.2, random_state=30)
```

```python
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier()
model.fit(X_train,y_train)
model.score(X_test,y_test)
```
