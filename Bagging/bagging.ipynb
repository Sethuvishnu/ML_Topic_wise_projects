```python
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv("heart.csv")
```

```python
data.info()
```

```python
data.describe()
```

```python
data1=data[data.Cholesterol<=data.Cholesterol.mean()+3*data.Cholesterol.std()]
```

```python
data1
```

```python
data2=data1[data1.Oldpeak<=data1.Oldpeak.mean()+3*data1.Oldpeak.std()]
```

```python
data2
```

```python
data3 = data2[data2.RestingBP<=(data2.RestingBP.mean()+3*data2.RestingBP.std())]
data4=data3.copy()
```

```python
data4.ST_Slope.replace(
    {
        'Down': 1,
        'Flat': 2,
        'Up': 3
    },
    inplace=True
)
```

```python
data4.RestingECG.replace(
    {
        'Normal': 1,
        'ST': 2,
        'LVH': 3
    },
    inplace=True)
```

```python
data4.ExerciseAngina.replace({'N':0,'Y':1},inplace=True)
```

```python
data5=pd.get_dummies(data4,drop_first=True)
```

```python
data5
```

```python
x=data5.drop('HeartDisease',axis='columns')
y=data5['HeartDisease']
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_=scaler.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(x_, y, test_size=0.2, random_state=10)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
from sklearn.model_selection import cross_val_score
```

```python
cross_val_score(DecisionTreeClassifier(),x_,y,cv=5)
```

```python
from sklearn.ensemble import BaggingClassifier
```

```python
bag_model=BaggingClassifier(base_estimator=DecisionTreeClassifier(),n_estimators=100,max_samples=0.8,oob_score=True)
```

```python
bag_model.fit(X_train,y_train)
```

```python
bag_model.oob_score_
```

```python
score=cross_val_score(bag_model,x_,y,cv=5)
score.mean()
```
