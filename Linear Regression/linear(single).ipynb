```python
import pandas as pd 
import numpy as np
from sklearn import linear_model
from word2number import w2n
```

```python
sal=pd.read_csv("hiring.csv")
```

```python
sal
```

```python
sal.fillna(0)
```

```python
sal['experience']=sal['experience'].replace(0,"zero")
```

```python
sal
```

```python
sal.experience = sal.experience.fillna("zero")
sal
```

```python
sal
```

```python
sal['experience'] = sal['experience'].fillna('zero')
sal.experience = sal.experience.apply(w2n.word_to_num)
sal
```

```python
sal.experience=sal.experience.fillna("zero").astype(str)

sal["experience"]=[w2n.word_to_num(x) for x in sal.experience]
```

```python
sal
```

```python
import math
```

```python
median=sal['test_score(out of 10)'].mean()
math.floor(median)
```

```python
sal["test_score(out of 10)"]=sal["test_score(out of 10)"].fillna(median)
sal
```

```python
reg=linear_model.LinearRegression()
reg.fit(sal[['experience','test_score(out of 10)','interview_score(out of 10)']],sal['salary($)'
])
```
